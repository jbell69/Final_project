```python
csvpath = 'paintings_master.csv'
csv_df = pd.read_csv(csvpath)
csv_df
```

```python
url_csv = 'consolidated.csv'
url_df = pd.read_csv(url_csv)
url_df.columns = ['Object_id', 'URL']
url_df
```

```python
merged = url_df.merge(csv_df, on="Object_id")
merged
```

```python
###Create dataframe for database
df = merged[['Object_id', 'Object_type', 'Title','Date_started','Date_finished','Medium', 'Classification','URL','Artist_id','Dept_id']]
###Rename columnst to match SQL database
df.columns = ['Object_id', 'Object_type', 'Title','Date_start','Date_finis','Medium', 'Classification','Met_link','Artist_id','Dept_id']
df
```

```python
###Save the progress 
df.to_csv('database.csv',index=False)
```
