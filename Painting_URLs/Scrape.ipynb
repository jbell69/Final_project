```python
###Dependency List and Additions
import sys
import os
import bs4
import requests
import pprint
import pandas as pd
from splinter import Browser
import csv
import time
browser = Browser('firefox', headless=True)
sys.modules.keys()
```

```python
###Use links so only scraping paintings
csvpath = 'paintings_master.csv'

links = []

with open(csvpath, newline="", encoding="utf8") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")

    # Loop through looking for the video
    for row in csvreader:
        links.append(row[1])
links
```

```python
### Scrape and put info in dictionaries
json_ls = []
json_di = {}
url = 'https://collectionapi.metmuseum.org/'
for i in links[links.index('637714'):]:
    query = 'public/collection/v1/objects/' + str(i)
    new_res = requests.get(url + query)
    new_resj = new_res.json()
    time.sleep(.1)
    try:
        img_link = new_resj['primaryImage']
        if img_link != '':
            json_ls.append(new_resj['objectID'])
            json_ls.append(new_resj['primaryImage'])
            json_di[new_resj['objectID']] = new_resj['primaryImage']
    except:
        pass
    print(query)
    print(img_link)
json_di
```

```python
###dictionary to dataframe
json_di
url_df = pd.DataFrame(list(zip(json_di.keys(),json_di.values())), columns=['Met ID', 'URL'])
url_df.tail()
```

```python
###Save the progress 
url_df.to_csv('637714_end.csv',index=False)
```
